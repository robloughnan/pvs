```python
def show_slices(slices):
    """ Function to display row of image slices """
    fig, axes = plt.subplots(1, len(slices))
    for i, slice in enumerate(slices):
        axes[i].imshow(slice.T, cmap="gray", origin="lower")
```

```python
slice_0 = tstats_data[90, :, :]
slice_1 = tstats_data[:, 100, :]
slice_2 = tstats_data[:, :, 70]
show_slices([slice_0, slice_1, slice_2])
plt.suptitle("Center slices for EPI image")  
```

```python
tstats = nib.load('./t_stats.nii')
tstats_data = tstats.get_fdata()
tmp = tstats_data.T
ind = abs(tstats_data[:, :, :].T)<3.5
tstats_data.T[ind] = np.nan
```

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 10))
alpha = 0.8
cm_stats = 'jet'

axial_slc = 40
axes[0].imshow(epi_img_data[:, :, axial_slc].T, cmap="gray", origin="lower")
axes[0].imshow(tstats_data[:, :, axial_slc].T, cmap=cm_stats, origin="lower", alpha=alpha)
axes[0].set_xticks([])
axes[0].set_yticks([])

axial_slc = 60
axes[1].imshow(epi_img_data[:, :, axial_slc].T, cmap="gray", origin="lower")
axes[1].imshow(tstats_data[:, :, axial_slc].T, cmap=cm_stats, origin="lower", alpha=alpha)
axes[1].set_xticks([])
axes[1].set_yticks([])

axial_slc = 70
axes[2].imshow(epi_img_data[:, :, axial_slc].T, cmap="gray", origin="lower")
axes[2].imshow(tstats_data[:, :, axial_slc].T, cmap=cm_stats, origin="lower", alpha=alpha)
axes[2].set_xticks([])
axes[2].set_yticks([])
```
